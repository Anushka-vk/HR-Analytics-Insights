```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_excel("../DataSets/HR data.xlsx")
df
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.replace(" ","")
```

```python
# Rename CF_ageband to AgeBand
renam = df.rename(columns={"CF_ageband":"AgeBand"})
renam
```

```python
#filter row where gender is male and attrition is yes
filt = df[(df.Gender == "Male") & (df.Attrition == "Yes")]
filt
```

```python
# New coloumn added (YearsSincePromotion)
df["YearsSincePromotion"] = df.YearsAtCompany - df.YearsSinceLastPromotion
```

```python
df
```

```python
# Group data by department and calculate average monthly income
df.groupby("Department")["MonthlyIncome"].mean()
```

```python
# Average monthly income for each Job Role in each Department
df.groupby(["Department","JobRole"])["MonthlyIncome"].mean().unstack()
```

```python
# Count of employee by gender and attrition
df.groupby(['Gender','Attrition'])['EmployeeCount'].count().unstack()
```

```python
# Total monthly Income by Gender and Department
df.groupby(['Gender','Department'])['MonthlyIncome'].sum().unstack()
```

```python
# convert Attrition column from object to numeric
df.Attrition = df.Attrition.map({'Yes':1,'No':0})
```

```python
# Proportion of employees by gender and attrition
pd.crosstab(index = df.Gender, columns=df.Attrition, margins=True)
```

```python
# correlation matrix for numerical columns
matrix = df.corr(numeric_only=True)
```
